```python
# Dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
A_df = pd.read_csv('countries-by-internet-users.csv')
A_df.head()
```

```python
A_df = A_df.rename(columns={"Country or Area":"Country"})
```

```python
A_df.head()
```

```python
B_df = pd.read_csv('world-happiness-report-2019.csv')
B_df.head()
```

```python
B_df = B_df.rename(columns={"Country (region)":"Country"})
```

```python
B_df.head()
```

```python
Combined_df = pd.merge(B_df, A_df, how='outer', on='Country')
```

```python
Combined_df.head()
```

```python
Combined_df = Combined_df.rename(columns={"SD of Ladder":"Rank of SD of Happiness Rank"})
Combined_df = Combined_df.rename(columns={"Ladder":"Happiness Rank"})
Combined_df = Combined_df.rename(columns={"Rank":"Internet Percent Usage Rank"})
Combined_df = Combined_df.rename(columns={"Log of GDP\nper capita":"Rank of GDP per Capita"})
Combined_df = Combined_df.rename(columns={"Healthy life\nexpectancy":"Rank of Health and Life Expectancy"})
Combined_df = Combined_df.rename(columns={"Generosity":"Rank of Generosity"})
Combined_df = Combined_df.rename(columns={"Corruption":"Rank of Corruption"})
Combined_df = Combined_df.rename(columns={"Freedom":"Rank of Freedom"})
Combined_df = Combined_df.rename(columns={"Social support":"Rank of Social Support"})
Combined_df = Combined_df.rename(columns={"Percentage":"Internet Users as Percentage of Population"})
del Combined_df["Rank.1"]
del Combined_df["Positive affect"]
del Combined_df["Negative affect"]
```

```python
Combined_df.head(250)
```

```python
Combined_df = Combined_df.dropna()
Combined_df.head()
```

```python
Combined_df['Internet and Happiness Score'] = Combined_df['Happiness Rank'] + Combined_df['Internet Percent Usage Rank']
Combined_df.head()
```

```python
Sorted_df = Combined_df.sort_values("Internet and Happiness Score")
Sorted_df.head()
```
